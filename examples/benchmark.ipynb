```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np

matplotlib.rcParams.update({'font.size': 8})
matplotlib.rcParams['figure.dpi'] = 300
```

```python
k_cpu_num= [20,40,64,128]
k_cpu_rate = [11.36,21.24,29.20,33.46]

k_gpu_num = [1,2,3]
k_gpu_rate = [340.96,340.72,343.93]

k_gpucpu_num = [1,4,10]
k_gpucpu_rate = [340.96,336.86,337.85]
#
fig, ax = plt.subplots(1,3, sharey=True)

ax[0].set_ylabel("Rate (ns/day)")
ax[0].set_xlim(0,150)
ax[0].set_ylim(0,420)
ax[0].plot(k_cpu_num,k_cpu_rate, color='tab:blue', marker='o')
ax[0].set_xticks(k_cpu_num)
ax[0].set_xlabel("CPU cores (0 GPUs)")

ax[1].plot(k_gpu_num,k_gpu_rate,  color='tab:blue', marker='o')
ax[1].set_xticks(k_gpu_num)
ax[1].set_xlabel("GPUs (1 CPU Core/GPU)")
ax[1].set_xlim(0,4)
ax[1].set_title("Benchmarking an 80k Atom System")

ax[2].plot(k_gpucpu_num,k_gpucpu_rate, color='tab:blue', marker='o')
ax[2].set_xticks(k_gpucpu_num)
ax[2].set_xlabel("CPU Cores (1 GPU)")
ax[2].set_xlim(0,11)



fig.set_size_inches(7.1,2)
fig.tight_layout()
fig.subplots_adjust(wspace=0)
fig.savefig('benchmark.png', format='png')
```
